## Word-By-Word model:

The idea here is to replace de char-by-char model by a word-by-word model. We want the NN to learn which word is better to follows.

For that porpose, let $W=\{w_i\}_{i=1}^{N}$ be the set of all the words in the corpus and $\texttt{idx_word}(w_i)=i$, $\texttt{word_idx}(i) = w_i$ the functions that represents the indexing of the set.

```python
r = requests.get("https://data.heatonresearch.com/data/t81-558/text/"\
                 "treasure_island.txt")

# raw_test is essentially a large string.
raw_text = r.text

text = raw_text.lower()
text = re.sub(r'[^\x00-\x7f]',r'', text)

W = text.split(' ')
```
